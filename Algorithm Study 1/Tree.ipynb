```python
class Node_Tree :
    def __init__(self, data) :
        self.data = data
        self.left, self.right = None, None
        
class BinarySearchTree :
    def __init__(self) :
        self.root = None
        self.len = 0
        
    def insert(self, data) :
        new_node = Node_Tree(data)
        if self.root == None :
            self.root = new_node
            
        else :
            current_node = self.root
            while True :
                if new_node.data < current_node.data :
                    if current_node.left != None :
                        current_node = current_node.left
                    else :
                        current_node.left = new_node
                        break
                        
                else :
                    if current_node.right != None :
                        current_node = current_node.right
                    else :
                        current_node.right = new_node
                        break
        self.len += 1
                        
    def search(self, data) :
        
        if self.root == None :
            return False
        else :
            current_node = self.root
            while True :
                if data == current_node.data :
                    return current_node
            
                else :
                    if data < current_node.data :
                        if current_node.left != None :
                            current_node = current_node.left
                        else :
                            return "Data Not Found"
                    
                    else :
                        if current_node.right != None :
                            current_node = current_node.right
                        else :
                            return "Data Not Found"
                        
    def delete(self, data) :
        if self.root == None :
            return False
        
        else :
            self.current_node = self.root
            self.parent_node = self.root
            while True :
                if self.current_node.data == data :
                    break
                    
                elif data < self.current_node.data :
                    if self.current_node.left != None :
                        self.parent_node = self.current_node
                        self.current_node = self.current_node.left
                        
                    else :
                        return "No Data to Delete"
                
                else :
                    if self.current_node.right != None :
                        self.parent_node = self.current_node
                        self.current_node = self.current_node.right
                    
                    else :
                        return "No Data to Delete"
                    
            
            if self.current_node.left == None and self.current_node.right == None:
                if self.current_node.data < self.parent_node.data :
                    self.parent_node.left = None
                                    
                else :
                    self.parent_node.right = None
                                    
                self.len -= 1
                    
            elif self.current_node.left != None and self.current_node.right == None :
                self.parent_node.right = self.current_node.left
                self.len -= 1
                
            elif self.current_node.left == None and self.current_node.right != None :
                self.parent_node.left = self.current_node.right
                self.len -= 1
                    
            else :
                current_node = self.current_node.right
                leaf_node_judgement = bool(current_node.left or current_node.right)
                
                while leaf_node_judgement :
                    if current_node.left != None :
                        break
                    
                    else :
                        current_node = current_node.left
                        
                if self.current_node.data < self.parent_node.data :
                    self.parent_node.left.data = current_node.data
                else :
                    self.parent_node.right.data = current_node.data
                    
                self.len -= 1
                                    
    def pre_order_traversal(self) :
        def _pre_order_traversal(root) :
            if root == None :
                pass
            else :
                print(root.data)
                _pre_order_traversal(root.left)
                _pre_order_traversal(root.right)
        _pre_order_traversal(self.root)
        
    def in_order_traversal(self) :
        def _in_order_traversal(root) :
            if root == None :
                pass
            else :
                _in_order_traversal(root.left)
                print(root.data)
                _in_order_traversal(root.right)
        _in_order_traversal(self.root)
        
    def post_order_traversal(self) :
        def _post_order_traversal(root) :
            if root == None :
                pass
            else :
                _post_order_traversal(root.left)
                _post_order_traversal(root.right)
                print(root.data)
        _post_order_traversal(self.root)
        
    def Size(self) :
        return self.len
        
        
        
                    
my = BinarySearchTree()
datum = [40, 4, 34, 45, 14, 55, 48, 13, 15, 49, 47]
for data in datum :
    my.insert(data)
    
my.delete(13)
my.post_order_traversal()
```
